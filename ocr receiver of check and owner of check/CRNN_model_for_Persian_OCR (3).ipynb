```python
# نصب کتابخانه hezar
!pip install hezar

# سایر کتابخانه‌های مورد نیاز
from hezar.models import Model
import os
from PIL import Image
import matplotlib.pyplot as plt

# بارگذاری مدل CRNN
crnn = Model.load("hezarai/crnn-base-fa-64x256")

# تابع برای خواندن و نمایش تصاویر و متون استخراج شده
def read_and_display_images(folder_path):
    # دریافت لیست فایل‌های تصویری در فولدر
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        # نمایش تصویر
        image = Image.open(image_path)
        plt.imshow(image)
        plt.axis('off')
        plt.show()

        # پیش‌بینی متن از تصویر
        texts = crnn.predict([image_path])
        print(f"متن استخراج شده از {image_file}: {texts[0]}")

# مسیر فولدر حاوی تصاویر
folder_path = '/content/image_folder'  # مسیر فولدر تصاویر خود را اینجا قرار دهید

# خواندن و نمایش تصاویر و متون استخراج شده
read_and_display_images(folder_path)
```
