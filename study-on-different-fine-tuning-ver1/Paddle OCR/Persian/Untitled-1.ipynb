```python
import pandas as pd

# Load the Excel file into a DataFrame using openpyxl
file_path = '/Users/hmi/documents/perset/ocr-image_to_txt.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1', engine='openpyxl')

# Remove the first row
df = df.iloc[1:].reset_index(drop=True)

# Save the resulting DataFrame to a CSV file with UTF-8 encoding
csv_file_path = '/Users/hmi/documents/ocr-image_to_txt.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
```

```python
def csv_to_txt_with_spacing(csv_file_path, txt_file_path):
    # Read the CSV file line by line
    with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
        lines = csvfile.readlines()

    # Open the TXT file for writing
    with open(txt_file_path, 'w', encoding='utf-8') as txtfile:
        for line in lines:
            # Strip leading/trailing whitespace and split on the first space
            line = line.strip()
            parts = line.split(maxsplit=1)
            
            # Ensure there are two parts after the split
            if len(parts) == 2:
                file_path = parts[0].strip()
                description = parts[1].strip()
                
                # Write to the TXT file with a tab separator
                txtfile.write(f"{file_path} {description}\n")

# Usage
csv_file_path = '/Users/hmi/documents/ocr-image_to_txt.csv'
txt_file_path = '/Users/hmi/Documents/pertune/val/output_label.txt'

csv_to_txt_with_spacing(csv_file_path, txt_file_path)

def replace_commas_with_tabs(input_file_path, output_file_path):
    # Read the text file line by line
    with open(input_file_path, 'r', encoding='utf-8') as infile:
        lines = infile.readlines()

    # Open the output file for writing
    with open(output_file_path, 'w', encoding='utf-8') as outfile:
        for line in lines:
            # Replace commas with tab spaces
            updated_line = line.replace(',', '        ')
            # Write the updated line to the output file
            outfile.write(updated_line)

# Usage
output_file_path = input_file_path = '/Users/hmi/Documents/pertune/val/output_label.txt'
replace_commas_with_tabs(input_file_path, output_file_path)
```

### rename files in with indexing 

```python
import os

def remove_spaces_in_filenames(folder_path):
    # Iterate over all files in the given folder
    for filename in os.listdir(folder_path):
        # Construct the full file path
        old_file_path = os.path.join(folder_path, filename)
        
        # Check if it's a file (not a directory)
        if os.path.isfile(old_file_path):
            # Remove spaces from the filename
            new_filename = filename.replace(" ", "")
            
            # Construct the new file path
            new_file_path = os.path.join(folder_path, new_filename)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: {old_file_path} -> {new_file_path}')

# Specify the folder path
folder_path = '/Users/hmi/Documents/perset'

# Call the function
remove_spaces_in_filenames(folder_path)
```
