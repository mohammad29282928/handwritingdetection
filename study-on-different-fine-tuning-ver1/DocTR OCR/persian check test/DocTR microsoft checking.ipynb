```python

import os
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Ensure the directory exists
os.makedirs("test1", exist_ok=True)

# List of detection architectures
det_architectures = [
    #'linknet_resnet18',
    #'linknet_resnet34',
    'linknet_resnet50',
    'db_resnet50',
    'fast_tiny',
    'fast_small',
    'fast_base',
    'db_mobilenet_v3_large',
    # 'db_mobilenet_v3_base',
    # 'db_mobilenet_v3_small'

]

# Path to the input folder
input_folder = "In"

# Get list of image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Loop through all image files
for idx, img_name in enumerate(image_files):
    # Rename images from 1 to n
    new_img_name = f"{idx+1}.jpg"
    os.rename(os.path.join(input_folder, img_name), os.path.join(input_folder, new_img_name))

    # Load the document
    image_path = os.path.join(input_folder, new_img_name)
    doc = DocumentFile.from_images(image_path)

    # Loop through all detection architectures and perform OCR
    for det_arch in det_architectures:
        print(f"Using detection architecture: {det_arch} on image: {new_img_name}")
        try:
            # Initialize the model
            model = ocr_predictor(det_arch=det_arch, reco_arch='crnn_vgg16_bn', pretrained=True)
            # Analyze the document
            result = model(doc)

            # Display the result
            result.show()

            filename = f"test1/{det_arch}_and_crnn_vgg16_bn_on_{new_img_name[:-4]}.txt"
            with open(filename, "w") as text_file:
                text_file.write(str(result))
                result.save(filename[:-4]+".png")
        except Exception as e:
            print(f"Error using detection architecture {det_arch} on image {new_img_name}: {e}")
```

```python
import zipfile
import os

def zip_directories(dirs, zip_filename):
    # Create a ZipFile object in write mode
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for dir_name in dirs:
            # Walk through the directory
            for root, dirs, files in os.walk(dir_name):
                for file in files:
                    # Create the full file path
                    file_path = os.path.join(root, file)
                    # Add file to the zip file, preserving directory structure
                    arcname = os.path.relpath(file_path, os.path.dirname(dir_name))
                    zipf.write(file_path, arcname=arcname)

# List of directories to zip
directories = ['test1', 'test2', 'test3', 'test4', 'test5', 'test6', 'test7', 'test8']

# Name of the output zip file
zip_filename = 'tests_archive.zip'

# Call the function to zip directories
zip_directories(directories, zip_filename)

print(f"Directories {directories} have been zipped into {zip_filename}")
```
